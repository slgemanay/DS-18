```python
import pandas as pd
import numpy
```

```python
df = pd.read_csv('ml-latest-small/ratings.csv')
df
```

```python
def calc(user_data):
    max_timestamp = max([int(x[3]) for x in user_data])
    min_timestamp = min([int(x[3]) for x in user_data])
    lifetime = max_timestamp - min_timestamp
    return lifetime
```

```python
user_average_lifetime = []
user_data = []
marks_counter = 0
previous_user = None
with open('ml-latest-small/ratings.csv') as f:
    f.readline()
    for line in f:
        userId, movieId, rating, timestamp = line.strip().split(',')
        if previous_user != None:
            if userId == previous_user:
                marks_counter +=1
                user_data.append([userId, movieId, rating, timestamp])
            else:
                previous_user = userId
                if marks_counter >= 100:
                    user_average_lifetime.append([previous_user, calc(user_data)])
                    marks_counter = 0
                else:
                    marks_counter = 0
        else:
            previous_user = userId
            user_data.append([userId, movieId, rating, timestamp])
            marks_counter = 1
user_average_lifetime
```

```python
def average_lifetime(user_average_lifetime):
    value_count = 0
    counter = 0
    counter = len(user_average_lifetime)
    value_count = sum([int(x[1]) for x in user_average_lifetime])
    answer = value_count/counter
    return answer
```

```python
average_lifetime(user_average_lifetime)
```
